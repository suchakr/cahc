```python
import pandas as pd
import numpy as np
import os
import re
from glob import glob
from pprint import pprint
import difflib as dl
```

```python
# read a file to a string
def read_file(file_path): 
	with open(file_path, 'r') as f: lines = f.read()
	lines = re.sub("[\n\r]+", " ", lines)
	lines = re.sub("\s+", " ", lines)
	lines = re.sub("([॥|।])\s", "\\1\n", lines)
	lines = re.split("\n", lines)
	print(len(lines))
	return lines




print ({ re.sub(".*\/","",f): read_file(f) for f in glob('../datasets/vgj-manuscripts-samples/*.txt') })

```
