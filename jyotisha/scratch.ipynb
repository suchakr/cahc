```python
# frequency count of words given a file
def freq_count(fn):
	with open(fn) as f: lines = f.readlines()
    	a0 = [ l.strip().split() for l in lines ]
    	a1 = [ item for sublist in a0 for item in sublist ]
    	a2 = [ re.sub(r'[^\w\s]','',item) for item in a1 ]
	freq_dict = { item:a1.count(item) for item in a1 }
	return freq_dict

# save dictionary to file
def save_dict(d,fn):
	with open(fn,'w') as f:
		for k,v in d.items():
			f.write(f'{k}\t{v}\n')

# load dictionary from file and save the dictionary
def load_dict(fn):
	src = fn
	tgt = fn.replace('.txt','.glossary')
	save_dict(freq_count(fn),tgt)
```
