## Monty Hall Problem - Would you like to switch or not?
- Behind three doors, one is a car, and the other two are goats. 
  - üö™ üö™ üö™ => üêê üêê üöó 
- You choose one of the doors and not open it.  
  - üö™ üö™ ‚úÖ => üêê üêê üöó 
- The host opens one of the doors, revealing a goat
  - üö™ ‚úîÔ∏é ‚úÖ => üêê üêê üöó 
- You can choose to switch or not switch
- Would you like to switch? ( assuming you like cars  )

````python
from random import randint
from numpy import array as npa
price_doors, initial_choices = ( 
	npa([randint(0, 2)
		for x in range (10000)
	] ) for y in range(2)
)
host_reveals = npa(
	[({0,1,2} - set([r]) - set([c])).pop() 
		for r,c in zip(price_doors,initial_choices
	)]
)
revised_choices = npa(
	[({0,1,2} - set([c]) - set([h])).pop() 
		for c,h in zip(initial_choices,host_reveals
	)]
)
print (	
f'''
Stick with Initial Choice {(price_doors == initial_choices).sum()/price_doors.size:.3f}% 
Switch the Initial Choice {(price_doors == revised_choices).sum()/price_doors.size:.3f}%
'''
)
````

````python
import pandas as pd
from random import randint
import math
import numpy as np
import matplotlib.pyplot as plt

def prep_grid_and_sensors(grid_size=20, num_sensors=5, sensor_placement="random"):
	cells = pd.Series(range(0,grid_size**2))

	sensor_placement = sensor_placement.lower()
	sensor_placement = "random" if sensor_placement == "random" else "chosen"

	corner_sensor_locations = [ 0 , grid_size-1 , grid_size//2 + (grid_size**2-1)//2 , grid_size*grid_size - grid_size, grid_size**2-1][:num_sensors]
	sensors =  cells.sample(n=num_sensors, replace=False) if sensor_placement == "random" else  cells[ cells.isin(corner_sensor_locations)]
	# print (f'Sensors are placed: {sensors.values}')
	print ( f'Processing Grid of {grid_size}x{grid_size}. '
		f'{sensors.shape[0]} {sensor_placement} Sensors(+) ' 
	)

	cell_sensor_distance = cells.apply( lambda c: sensors.apply(lambda s: (s//grid_size-c//grid_size)**2 +(c%grid_size-s%grid_size)**2)**0.5)
	# display("Distance of each cell from the sensor", pd.concat([cell_sensor_distance.head(2), cell_sensor_distance.tail(2)]) )
	#cell_sensor_distance_for_cells_with_no_sensors=cell_sensor_distance[[ix  not in sensors.index for ix in cell_sensor_distance.index]]
	return cell_sensor_distance, sensors
	
def simulate_shoot_and_detect(grid_size=20, num_sensors=5, sensor_placement="random"):
	cell_sensor_distance, sensors = prep_grid_and_sensors(grid_size, num_sensors, sensor_placement)

	fig, axs = plt.subplots(nrows=1 , ncols=4, figsize=(5*4,4), sharex=True, sharey=True, )
	fig.suptitle(
		f'Grid of {grid_size}x{grid_size}. '
		f'{sensors.shape[0]} {sensor_placement} Sensors(+) ' 
		f' -- ' 
		f'Star is Actual Shot. ' 
		f'Circle is Computed Shot. ' 
		, fontsize=16
	)
	fig.subplots_adjust(top=0.85)
	for ix, ax in zip(range(4), axs.flatten()):
		#simulated_shot_no_noise = cell_sensor_distance_for_cells_with_no_sensors.sample(1)
		simulated_shot_no_noise = cell_sensor_distance.sample(1)
		# display(f'Simulated Shot no noise at {simulated_shot_no_noise.index[0]}', pd.DataFrame(simulated_shot_no_noise))
		simulated_shot = simulated_shot_no_noise + [3*randint(-9999,9999)/10000 for x in sensors]
		simulated_shot = simulated_shot.apply( lambda x: x - 1*x.min(), axis=1)	
		# display(f'Simulated Shot with noise at {simulated_shot.index[0]}', pd.DataFrame(simulated_shot))

		fit_attempt = cell_sensor_distance.apply( lambda x: ((simulated_shot-(x-1*x.min()))**2).sum()**.5, axis=1)
		fit_attempt['error'] = fit_attempt.apply(lambda x: x.sum(), axis=1)
		fit_attempt['row'] = fit_attempt.apply(lambda x: x.name//grid_size, axis=1)
		fit_attempt['col'] = fit_attempt.apply(lambda x: x.name%grid_size, axis=1)

		ax=fit_attempt.plot.scatter(x='col', y='row'#, c='error'
			# , figsize=(4,4)
			, marker='o'
			, title=f'Actual- {(simulated_shot.index[0]//grid_size,simulated_shot.index[0]%grid_size)} ~ ' 
				f'Computed - {(fit_attempt.error.idxmin()//grid_size,fit_attempt.error.idxmin()%grid_size)}'
			, s=fit_attempt.error.apply(lambda x: x*10)
			, colorbar=False
			, alpha=0.2
			, c= 'orange' if sensor_placement == 'random' else 'cyan'
			, ax=ax
		)

		x = np.array([simulated_shot.index[0]%grid_size])
		y = np.array([simulated_shot.index[0]//grid_size])
		ax.scatter(x, y, s=[300], c=['purple'], alpha=1, marker='*')

		x = np.array([fit_attempt.error.idxmin()%grid_size])
		y = np.array([fit_attempt.error.idxmin()//grid_size])
		ax.scatter(x, y, s=[200], c=['blue'], alpha=.3, marker='o')

		x = np.array([sensors.index%grid_size])
		y = np.array([sensors.index//grid_size])
		ax.scatter(x, y, s=[200], c=['black'], alpha=1, marker='P')

# simulate_shoot_and_detect(num_sensors=0, gridsize=20)
# simulate_shoot_and_detect(num_sensors=0, gridsize=30)

for sensors in range(4,0,-1):
	# simulate_shoot_and_detect(num_sensors=sensors, grid_size=20, sensor_placement="chosen")
	simulate_shoot_and_detect(num_sensors=sensors, grid_size=20, sensor_placement="random")

# for sensors in range(3,0,-1):
# 	simulate_shoot_and_detect(num_sensors=sensors, grid_size=30, sensor_placement="chosen")
# 	simulate_shoot_and_detect(num_sensors=sensors, grid_size=30, sensor_placement="random")
````

````python
from sympy import *
from pprint import pprint
x, y, t = symbols('x y t', positive=True)

grid_size=30
cells, sensors = prep_grid_and_sensors(grid_size=grid_size, num_sensors=3, sensor_placement="chosen")
sensor_xs, sensor_ys = sensors.index.values//grid_size, sensors.index.values%grid_size
non_sensor_cells = cells.drop(sensors.index)

for _ in range(5):
	shot_actual_duration = non_sensor_cells.sample(n=1) # assumin sound velocity is 1
	sensor_ts = shot_actual_duration.apply(lambda x: x-x.min(),axis=1)


	eqns = [ # solving these equations will give (x, y - location of shot) and  (t - time of shot to reach first receiver)
		(x- _x)**2 + (y-_y)**2 - (t + int(_t*1)/1.0)**2  
		for _x, _y, _t in zip(sensor_xs, sensor_ys, sensor_ts.values[0])
	]

	display("Input Time Profile based on a shot", sensor_ts)

	print('find x ,y and t by solving these equations\n', eqns )
	ans = pd.DataFrame(solve(eqns))
	if ( len(ans) == 0 ):
		print('no solution found')
		print("**********************************************************")
		continue
	ans['actual_x'] = shot_actual_duration.index.values[0]//grid_size
	ans['actual_y'] = shot_actual_duration.index.values[0]%grid_size
	ans.columns = ['t(solved_time_to_first_sensor)', 'x(solved)', 'y(solved)', 'actual_x', 'actual_y']
	ans = ans.applymap(lambda x: round(x,2))
	display(ans)
	print("===========================================================")
````

- Let $(x,y,t)$ represent the location of gunshot and the time the shot was fired - these are the **unknowns**
- Let $(X,Y,T)_i$ be the location and time of the shot heard by the $i^{th}$ sensor - these are the knowns

---
- $V_i = \frac{\sqrt{(X_i -x)^2 + (Y_i-y)^2}}{T_i - t}$ is the velocity of sound the $i^th$ sensor - We know all $V_is$ are identical
---
- $\forall{i}$   ${V_i = V_1}$ , yields ${i}$ quadratic? equations with three unkowns ${x,y,t}$
- Solving this system of equations for the real roots of $x,y,t$ gives us the location and time of the gunshot
- Since there are three unknows, three independent equations, hence three sensors are needed.
